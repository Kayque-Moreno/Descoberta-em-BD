```python
dados_usados = list(colecao.find({
    'valor' : {'$gte' : 90, '$lte' : 200}
}))
display(dados_usados)
```

```python
dados_por_cidade = list(colecao.find({
    'cidade' : {'$in' : ['Assis', 'Ourinhos', 'Pompeia', 'Marilia']}
}))

display(dados_por_cidade)
```

```python
df = pd.DataFrame(dados_por_cidade)
df
```

```python
df.isnull().sum()
```

```python
faturamento_total = df['valor'].sum()
print ("O faturamento total foi de: ", faturamento_total)
```

```python
faturamento_por_cidade = df.groupby('cidade')['valor'].sum().sort_values(ascending = True)
faturamento_por_cidade
```

```python
faturamento_por_servico = df.groupby('servico')['valor'].sum().sort_values(ascending = True)
faturamento_por_servico
```

```python
ticket_medio_geral = df['valor'].mean()
print(f"Ticket Médio Geral: {ticket_medio_geral:.2f}")
```

```python
ticket_medio_por_cidade = df.groupby('cidade')['valor'].mean().round(2)
print("Ticket Médio por Cidade: ", ticket_medio_por_cidade)
```

```python
servico_top = df.groupby('servico')['valor'].sum().idxmax()
print("O serviço mais lucrativo foi: ", servico_top)
```

```python
cidade_top = df.groupby('cidade')['valor'].sum().idxmax()
print("A cidade com maior faturamento foi: ", cidade_top)
```

```python
maiores_servicos = df.groupby('servico')['valor'].sum().sort_values(ascending = False)
maiores_servicos.head(3)
```
