```python
dados_mongo = list(colecao.find({
    'preco_unitario' : {'$gte' : 100, '$lte' : 300}
}))

display(dados_mongo)
```

```python
dados_cidade = list(colecao.find({
    'cidade' : {'$in' : ['Ourinhos', 'Assis']}
}))

display(dados_cidade)
```

```python
df = pd.DataFrame(dados_cidade)
df
```

```python
df.isnull().sum()
```

```python
df['valor_total'] = df['quantidade' ] * df['preco_unitario'] #criou uma coluna chamada VALOR_TOTAL e calculou o total de cada compra
df.head(3)
```

```python
faturamento = df['valor_total'].sum() #ver o faturamento total de todos os itens
print("Faturamento Total: ",  faturamento)
```

```python
faturamento_categoria = df.groupby('categoria')['valor_total'].sum().sort_values(ascending = False) #agrupa por CATEGORIA e vê quanto cada uma faturou; organiza do maior p/ o menor
faturamento_categoria
```

```python
faturamento_produto = df.groupby('produto')['valor_total'].sum().sort_values(ascending = False) #agrupa por PRODUTO e vê quanto cada um faturou; organiza do maior p/ o menor
faturamento_produto
```

```python
vendas_por_cidade = df['cidade'].value_counts() #ver qts vendas tem em cada CIDADE
vendas_por_cidade
```

```python
media_valor_vendas = df['valor_total'].mean() #calcular a média das vendas
print(f"Média do valor das vendas: R${media_valor_vendas:.2f}" )
```

```python
produto_top = df.groupby('produto')['valor_total'].sum().idxmax() #ver produto mais vendido
print("Produto mais vendido:" , produto_top)
```

```python
cidade_top = df.groupby('cidade')['valor_total'].sum().idxmax() #ver cidade com mais vendas
print ("Cidade com mais vendas: ", cidade_top)
```

```python
relatorio = {
    'data_geracao' : datetime.now(),
    'faturamento_total' : float(faturamento),
    'faturamento_categoria' : faturamento_categoria.to_dict(),
    'produto_top' : produto_top,
    'cidade_top' : cidade_top 
}
display(relatorio)
```

```python
colecao_relatorios = db['relatorios']
colecao_relatorios.insert_one(relatorio)
```
